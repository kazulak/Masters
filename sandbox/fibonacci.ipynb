```python
# Calculate the n-th Fibonacci number using an optimized approach
def fibonacci(n):
    if n < 0:
        raise ValueError("Negative arguments not implemented")
    return _fib(n)[0]

def _fib(n):
    if n == 0:
        return (0, 1)
    a, b = _fib(n >> 1)  # Recursive call for n//2
    c = a * (2*b - a)
    d = a*a + b*b
    if n & 1:
        return (d, c + d)
    else:
        return (c, d)
        
# Example usage
print(fibonacci(100))  # Output: 55
```
